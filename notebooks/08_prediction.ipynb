```python
with open(FEATURE_CLUSTER_PATH,'r') as aa:
    clusters=yaml.safe_load(aa)
clusters    
```

```python
aa={}
for cluster,skills in clusters.items():
    print(cluster,skills)
    for skill in skills:
        aa[skill]=cluster
```

```python
molten_clusters = [(cluster_name, cluster_skill)
                   for cluster_name, cluster_skills in clusters.items()
                   for cluster_skill in cluster_skills]

clusters_df = pd.DataFrame(molten_clusters, columns=["cluster_name", "skill"])
clusters_df
```

```python
group=['Python','SQL','Pandas','Keras','TensorFlow']
clusters_df["exist"]=clusters_df["skill"].isin(group)
clusters_df
```

```python
cluster_features=clusters_df.groupby("cluster_name")['exist'].sum()
```

```python
cluster_features
```

```python
cluster_features.index
```

```python

# skills_features=aa['exist']
# skills_features.index=aa["skills"]
# skills_features
```

```python
skills_features=features[~features.isin(cluster_features.index)]
skills_features
```

```python
group=['Python','SQL','Pandas','Keras','TensorFlow']
```

```python
ohe_skills = pd.Series(skills_features.isin(group).astype(int).tolist(), 
                       index=skills_features)
ohe_skills
```

```python
all_features=pd.concat([ohe_skills,cluster_features])
all_features
```

```python
predictions = classifier.predict_proba([all_features.values])
predictions
```

```python
positive_probs = [prob[0][1] for prob in predictions]
positive_probs
```

```python
result=pd.Series(positive_probs,index=targets)
result
```
